We're interested in the cross reactivity of designed HCV3A oligonucleotides designed by Irene.  
The oligonucleotides will be ran under the assumption of 'standard' PCR conditions and methods.

The following data was obtained from a series of steps:  
1) Ran the oligonucleotides in GenomeQuest using the NCBI BLAST algorithm.  
2) Ran the output of GenomeQuest with Tongjia's Proximity Tool with maxLength 3000  
3) Ran Derek's proximityFilter to check for correct orientation.  
4) Ran David's proximityrip to extract amplified hits.  


```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
fig = plt.figure(figsize=(14,12))
ax = fig.add_subplot(221)

with open('results/ndb_human-fin.tsv') as f:
    
    tsv_reader = csv.reader(f, delimiter='\t')
    
    forward_oligo_names = [for_name[3] for for_name in tsv_reader]
    f.seek(0)
    forward_oligo_query = [query_id[6] for query_id in tsv_reader]
    f.seek(0)
    reverse_oligo_names = [rev_name[4] for rev_name in tsv_reader]
    f.seek(0)
    reverse_oligo_query = [rev_query_id[10] for rev_query_id in tsv_reader]
    f.seek(0)
    subject_names_match = [subject_name[2] for subject_name in tsv_reader]
    f.seek(0)
    subject_e_value = [e_val[5] for e_val in tsv_reader]
    f.seek(0)
    
    forward_oligo_names.pop(0)
    forward_oligo_query.pop(0)
    reverse_oligo_names.pop(0)
    reverse_oligo_query.pop(0)
    subject_names_match.pop(0)
    subject_e_value.pop(0)
    
    ax.set_xlim([0,100])
    ax.set_ylim([0,100])
    plt.plot(forward_oligo_query, reverse_oligo_query, "x")
    plt.plot([0,100],[90,90], "r--")
    plt.plot([90,90],[0,100], "r--")
    plt.annotate("90% Match Zone", (93,93))
    plt.title("NDB-HUMAN")
    plt.xlabel("Query ID%\nFORWARD OLIGO")
    plt.ylabel("Query ID%\nREVERSE OLIGO")
    
ay = fig.add_subplot(222)

with open('results/sdb-fin.tsv') as f:
        
    tsv_reader = csv.reader(f, delimiter='\t')
    
    forward_oligo_names = [for_name[3] for for_name in tsv_reader]
    f.seek(0)
    forward_oligo_query = [query_id[6] for query_id in tsv_reader]
    f.seek(0)
    reverse_oligo_names = [rev_name[4] for rev_name in tsv_reader]
    f.seek(0)
    reverse_oligo_query = [rev_query_id[10] for rev_query_id in tsv_reader]
    f.seek(0)
    subject_names_match = [subject_name[2] for subject_name in tsv_reader]
    f.seek(0)
    subject_e_value = [e_val[5] for e_val in tsv_reader]
    f.seek(0)
    
    forward_oligo_names.pop(0)
    forward_oligo_query.pop(0)
    reverse_oligo_names.pop(0)
    reverse_oligo_query.pop(0)
    subject_names_match.pop(0)
    subject_e_value.pop(0)
    
    # Create a separate plot for strains that are strain 3a.
    indexes_of_strain = [num for num, x in enumerate(subject_names_match) if x.find("3a") > -1]
    indexes_of_forward = [forward_oligo_query[for_index] for for_index in indexes_of_strain]
    indexes_of_reverse = [reverse_oligo_query[rev_index] for rev_index in indexes_of_strain]
    
    
    ax.set_xlim([0,100])
    ax.set_ylim([0,100])
    plt.plot(indexes_of_forward, indexes_of_reverse, "rD")
    plt.plot(forward_oligo_query, reverse_oligo_query, "x")
    plt.plot([0,100],[90,90], "r--")
    plt.plot([90,90],[0,100], "r--")
    plt.annotate("90% Match Zone", (93,93))
    plt.title("Hepatitis C (SDB)")
    plt.xlabel("Query ID%\nFORWARD OLIGO")
    plt.ylabel("Query ID%\nREVERSE OLIGO")
    
az = fig.add_subplot(223)
    
with open('results/ndb-viral-fin.tsv') as f:
        
    tsv_reader = csv.reader(f, delimiter='\t')
    
    forward_oligo_names = [for_name[3] for for_name in tsv_reader]
    f.seek(0)
    forward_oligo_query = [query_id[6] for query_id in tsv_reader]
    f.seek(0)
    reverse_oligo_names = [rev_name[4] for rev_name in tsv_reader]
    f.seek(0)
    reverse_oligo_query = [rev_query_id[10] for rev_query_id in tsv_reader]
    f.seek(0)
    subject_names_match = [subject_name[2] for subject_name in tsv_reader]
    f.seek(0)
    subject_e_value = [e_val[5] for e_val in tsv_reader]
    f.seek(0)
    
    forward_oligo_names.pop(0)
    forward_oligo_query.pop(0)
    reverse_oligo_names.pop(0)
    reverse_oligo_query.pop(0)
    subject_names_match.pop(0)
    subject_e_value.pop(0)
    
    ax.set_xlim([0,100])
    ax.set_ylim([0,100])
    plt.plot(forward_oligo_query, reverse_oligo_query, "x")
    plt.plot([0,100],[90,90], "r--")
    plt.plot([90,90],[0,100], "r--")
    plt.annotate("90% Match Zone", (93,93))
    plt.title("NDB-VIRAL")
    plt.xlabel("Query ID%\nFORWARD OLIGO")
    plt.ylabel("Query ID%\nREVERSE OLIGO")

aa = fig.add_subplot(224)    

with open('results/ndb_invert-plant-bact-fin.tsv') as f:
        
    tsv_reader = csv.reader(f, delimiter='\t')
    
    forward_oligo_names = [for_name[3] for for_name in tsv_reader]
    f.seek(0)
    forward_oligo_query = [query_id[6] for query_id in tsv_reader]
    f.seek(0)
    reverse_oligo_names = [rev_name[4] for rev_name in tsv_reader]
    f.seek(0)
    reverse_oligo_query = [rev_query_id[10] for rev_query_id in tsv_reader]
    f.seek(0)
    subject_names_match = [subject_name[2] for subject_name in tsv_reader]
    f.seek(0)
    subject_e_value = [e_val[5] for e_val in tsv_reader]
    f.seek(0)
    
    forward_oligo_names.pop(0)
    forward_oligo_query.pop(0)
    reverse_oligo_names.pop(0)
    reverse_oligo_query.pop(0)
    subject_names_match.pop(0)
    subject_e_value.pop(0)
    
    ax.set_xlim([0,100])
    ax.set_ylim([0,100])
    plt.plot(forward_oligo_query, reverse_oligo_query, "x")
    plt.plot([0,100],[90,90], "r--")
    plt.plot([90,90],[0,100], "r--")
    plt.annotate("90% Match Zone", (93,93))
    plt.title("NDB-INVERT-BACT-PLANT")
    plt.xlabel("Query ID%\nFORWARD OLIGO")
    plt.ylabel("Query ID%\nREVERSE OLIGO")
```

NDB(Non-specific Database) contains the subjects we do not want to match with our oligonucleotide system. SDB(Specific Database) contains the subjects we do want to match with our oligonucleotide systems. Our goal is to see if our oligonucleotides chosen will be specific to Hepatitis C, mainly Hep C strain 3A.  

The red diamonds shown in the SDB graph represents strains that are targeted as Hep C strain 3A. The majority of the 3a strains had mostly approximately ~90% match to our oligonucleotide systems. Although our system also picks up other strains of Hepatitis C, we encompass the majority of what we intend to target.
